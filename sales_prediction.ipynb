```python
# Importando as bibliotecas necessárias.
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
# Carregando a base de dados.
dataset = pd.read_csv('C:/Users/raiss/dados/historico_vendas_coatings V2.csv', sep = ';', encoding='utf-8')
```

```python
# Visualizando os dados.
dataset.head().T
```

```python
#Verificando categoria dos dados
dataset.dtypes
```

```python
# Para os testes Iniciais selecionei apenas 3 atributos.
dataset[['Material Group', 'Invoice Price BRL /KG','Actual Billed Qty /KG']][:5]
```

```python
#Variáveis de Entrada.
X = dataset[['Material Group', 'Invoice Price BRL /KG']]

#Variável de Saída.
Y = dataset['Actual Billed Qty /KG']
```

```python
# 70% dos dados para treino e 30% para teste.
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
# Tamanho dos dados de treino.
x_train.shape, y_train.shape
```

```python
# Tamanho dos dados de teste.
x_test.shape, y_test.shape
```

```python
# Iniciando a Regressão
lreg = LinearRegression()
```

```python
# Treinando o Modelo
lreg.fit(x_train,y_train)
```

```python
# Predizendo os valores para o conjunto de teste
pred = lreg.predict(x_test)
```

```python
# Calculando o MSE - Mean Square Error
mse = np.mean((pred - y_test)**2)
mse
```

```python
# Calculando os Coeficientes
coeff = DataFrame(x_train.columns)
coeff['Coeficientes'] = Series(lreg.coef_)
coeff
```

```python
#Calculando o R-Squared
lreg.score(x_test,y_test)
```
